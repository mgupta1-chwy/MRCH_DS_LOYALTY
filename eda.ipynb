```python
metrics = []

def display_trended_graphs(dataset):
    
    dataset.sort_values(by = "SESSION_DATE" , inplace = True)
    
    tempPlot=dataset.copy()
    tempPlot.drop_duplicates(keep='first',inplace=True)
    data_v2 = dataset.copy()
    data_v2 = data_v2.iloc[:-1 , :]

#     data_v2=dataset[(dataset["DATE"]<=tempPlot["endDate"][0]) ]

    fig = make_subplots(
                rows=len(metrics), cols=1,
                subplot_titles=(metrics),
                shared_xaxes=True,
                vertical_spacing =0.04)

    for i in range(len(metrics)):
        fig.add_trace(go.Scatter(x=dataset["DATE"], y= data_v2[metrics[i]], mode='lines' , name = metrics[i]), row=i+1, col=1 )

    for index, row in tempPlot.iterrows():
        fig.add_vrect(x0=row["startDate"], x1=row["endDate"],
                      fillcolor=row["color"], opacity=0.15, line_width=0)
    fig.update_layout(width=1000,height=250*len(metrics),
                     yaxis2=dict())
    fig.show()
```
